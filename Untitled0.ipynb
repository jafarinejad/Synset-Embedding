```python
def med_synemb (w):
    synemb = np.zeros((50,))
    num = 0
    for fwsyn in wn.synsets(w):
        synemb += loaded_model.wv[fwsyn.name()]
        num +=1
    if num>0:
        synemb /=num
    #print( synemb)
    return synemb
```

```python
def to50(emb):
    res = np.zeros((50,))
    cc = 6#len(emb)/50
    for i in range(50):
        #print((i+1)*cc)
        res[i] = np.mean(emb[i*cc:(i+1)*cc])
    return(res)

fff = np.zeros((300,))
fff[10]=6
print(to50(fff))
            
```
